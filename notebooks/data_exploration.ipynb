### Correlation Matrix and Removing Redundant Features

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute the correlation matrix
corr_matrix = merged_df.corr()

# Set up the figure size
plt.figure(figsize=(15, 12))  # Adjusted for better visibility

# Create the heatmap with smaller font size
sns.heatmap(corr_matrix, cmap="coolwarm", annot=False, linewidths=0.5, square=True,
            xticklabels=True, yticklabels=True)

# Adjust the font size for better readability
plt.xticks(fontsize=8, rotation=90)  # Rotate feature names for clarity
plt.yticks(fontsize=8)

# Add title
plt.title("Correlation Matrix", fontsize=14)

# Show the plot
plt.show()
```

```python
# Check strongly correlated feature pairs
import numpy as np

# Get pairs with absolute correlation â‰¥ 0.9
high_corr_pairs = (
    corr_matrix.where(np.abs(corr_matrix) >= 0.9)  # Filter by threshold
    .stack()  # Convert to pairs
    .reset_index()
)

# Rename columns
high_corr_pairs.columns = ["Feature_1", "Feature_2", "Correlation"]

# Remove self-correlations (correlation of 1 with itself)
high_corr_pairs = high_corr_pairs[high_corr_pairs["Feature_1"] != high_corr_pairs["Feature_2"]]

# Sort by absolute correlation value
high_corr_pairs = high_corr_pairs.sort_values(by="Correlation", ascending=False)

print(high_corr_pairs)
```

```python
import pandas as pd

condition_columns = [
    "CONDITION_BnkL", "CONDITION_BnkR", "CONDITION_CS",
    "CONDITION_FE", "CONDITION_GR", "CONDITION_SlpD", "CONDITION_SlpU"
]

# Function to convert boolean values into numeric
def convert_and_save(csv_filename):
    # Load the CSV file
    df = pd.read_csv(csv_filename)
    
    # Convert boolean values to numerical (1 and 0) if the columns exist in the dataframe
    for col in condition_columns:
        if col in df.columns:
            df[col] = df[col].astype(int)
    
    new_filename = csv_filename.replace(".csv", "_numeric.csv")
    df.to_csv(new_filename, index=False)
    print(f"Updated dataframe saved as '{new_filename}'.")

convert_and_save("Merged_Data_Selected_Features.csv")
convert_and_save("Merged_Data_OneHotEncoded.csv")
```

### Treating Outliers using IQR Clipping

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/Merged_Data_OneHotEncoded_numeric.csv")

feature_columns = df.columns[:69]

# Compute Q1, Q3, and IQR
Q1 = df[feature_columns].quantile(0.25)
Q3 = df[feature_columns].quantile(0.75)
IQR = Q3 - Q1  # Compute IQR

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Apply IQR Clipping
df_clipped = df.copy()
df_clipped[feature_columns] = df[feature_columns].clip(lower=lower_bound, upper=upper_bound, axis=1)
```

```python
# Plot outliers after clipping

def plot_outliers(df, columns):
    for col in columns:
        plt.figure(figsize=(8, 4))
        sns.boxplot(x=df[col])
        plt.title(f"Boxplot of {col}")
        plt.show()
        
print("Outliers After IQR Clipping:")
plot_outliers(df_clipped, feature_columns)
```

```python
# Save the clipped dataframe to CSV file
df_clipped.to_csv("../data/Merged_Dataset_OneHotEncoded_numeric_IQR_Clipped.csv", index=False)
print("IQR-clipped dataset saved successfully")
```
